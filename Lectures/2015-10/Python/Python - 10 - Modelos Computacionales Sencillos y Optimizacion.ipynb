```python
def energy():
    cuenta=0
    for i in range(L):
        for j in range(L):
            i1=mod(i+1,L)
            j1=mod(j+1,L)
            i2=mod(i-1,L)
            j2=mod(j-1,L)
            s0=red[i][j]
            s1=red[i1][j]
            s2=red[i][j1]
            s3=red[i2][j]
            s4=red[i][j2]
            cuenta+=s0*(s1+s2+s3+s4)
    return -cuenta
def magnetization():
    cuenta = 0
    for j in red.flatten():
        if j == 1:
            cuenta+=1
        else:
            cuenta+=(-1)
    return cuenta
def ising():
    # Elegir una posición aleatoria en la red
    i=random.randint(0,L)
    j=random.randint(0,L)
    # Calcular la energía del bloque
    i1=mod(i+1,L)
    j1=mod(j+1,L)
    i2=mod(i-1,L)
    j2=mod(j-1,L)
    s0=red[i][j]
    s1=red[i1][j]
    s2=red[i][j1]
    s3=red[i2][j]
    s4=red[i][j2]
    if(s0==1):
        s5=-1
    else:
        s5=1
    energiai=-s0*(s1+s2+s3+s4)
    energiaf=-s5*(s1+s2+s3+s4)
    if energiaf<energiai:
        red[i][j]=s5
    else:
        # Lanzar un número aleatorio
        if(random.random()<exp(-1*(energiaf-energiai)/T)):
            red[i][j]=s5
    return 0
```

```python
# Modelo de Ising con el algoritmo de Metrópolis
L=64 # Lado de la red 
iter=100*L*L # Número de iteraciones
nummag=100 # Número de iteraciones para calcular la magnetización
longhistory=[]
longdevs=[]
T=0.
temps=[]
# Inicializar el arreglo
global red
red=zeros((L,L))
# Iterar sobre diferentes temperaturas
for i in range(50):
    print i,
    T+=0.1
    historia=[]
    counter=0
    while counter < iter:
        ising()
        if counter >= iter-nummag:
            historia.append(magnetization())
        counter+=1
    temps.append(T)
    longhistory.append(mean(historia))
    longdevs.append(std(historia))
```

```python
# Matriz de distancias entre ciudades
dists=genfromtxt("/Users/phoenix/Dropbox/Data/coldists.csv",delimiter=",")
# Latitud y longitud de las ciudades
coords=genfromtxt("/Users/phoenix/Dropbox/Data/citycoords.csv",delimiter=",")
```

```python
# En total son 27 ciudades
# Bogotá está en el primer renglón
# Comenzar siempre en Bogotá y terminar en Bogotá
# La función genera un par de números enteros en el intervalo [1,26]
def cambio():
    primero=random.randint(1,26)
    persiste=1
    while persiste==1:
        segundo=random.randint(1,26)
        if segundo != primero:
            persiste=0
    return [primero,segundo]
```

```python
# Las siguientes dos líneas inicializan el camino
# siguiendo las ciudades en el orden original
camino=range(1,28)
camino[-1]=1
thelats = coords[:,1]
thelons = coords[:,2]
lats=[]
lons=[]
for i in camino:
    lats.append(thelats[i])
    lons.append(thelons[i])
fig = plt.figure(figsize=(10,10))
ax = fig.add_axes([0,0,1,1])
m = Basemap(llcrnrlon=-80,llcrnrlat=-4.,urcrnrlon=-60,urcrnrlat=13,resolution='l',\
            area_thresh=1000.,projection='lcc',lat_1=1,lon_0=-2.,ax=ax)
x, y = m(lons,lats)
m.plot(x,y,color='w')
m.plot(x,y,"o",color='w')
m.drawmapboundary(fill_color='#99ffff')
m.drawcountries()
m.fillcontinents(color='#cc9966',lake_color='#99ffff')
plt.show()

dist=0
for k in xrange(26):
    city1=camino[k]
    city2=camino[k+1]
    dist+=dists[city1-1][city2]
print dist," km"
```

```python
# Implementación de un algoritmo genético
camino=range(1,28)
camino[-1]=1
intentos=1000000
historia=[]
for i in xrange(intentos):
    #calcular la distancia total recorrida
    dist=0
    if i%100000==0:
        print i,
    for k in xrange(26):
        city1=camino[k]
        city2=camino[k+1]
        dist+=dists[city1-1][city2]
    historia.append(dist)
    #Hacer un cambio en el camino
    mole=cambio()
    camino[mole[0]], camino[mole[1]] = camino[mole[1]], camino[mole[0]]
    # Calcular la nueva distancia
    newdist=0
    for k in xrange(26):
        city1=camino[k]
        city2=camino[k+1]
        newdist+=dists[city1-1][city2]
    # Si la nueva distancia es menor entonces conservar el cambio de lo contrario revertirlo
    if newdist > dist:
        camino[mole[0]], camino[mole[1]] = camino[mole[1]], camino[mole[0]]
min(historia)
```

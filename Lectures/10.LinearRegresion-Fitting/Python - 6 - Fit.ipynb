# Ley de Hubble

```python
distances = galaxies[:,1] # En kPc
speeds = galaxies[:,2] # En km/s
distances = distances / 1000. # En MPc
# Antes de hacer la gráfica calculemos los parámetros de la regresión lineal
thefit=polyfit(distances,speeds,1)
slope=thefit[0]
intercept=thefit[1]
```

```python
print thefit
```

```python
#Ejemplo de regresión lineal
figure(figsize=(10,5))
plot(distances,speeds,"o",ms=1)
plot([0,160],[intercept,160*slope+intercept],'c--')
title(u'Ley de Hubble\nEjemplo de regresión lineal',fontsize=20)
xlabel('$\mathrm{distancia}(\mathrm{MPc})$',fontsize=15)
ylabel(r'$\mathrm{rapidez}(\frac{km}{s})$',fontsize=15)
ylim(0)
show()
```

## Movimiento parabólico

$y(x)=\frac{v_{i,y}}{v_{i,x}}x-\frac{1}{2}g\frac{x^2}{v_{i,x}^2}$

```python
# Ejemplo de regresión cuadrática
flight=genfromtxt(os.path.expanduser("~/Dropbox/Data/flight.csv"),delimiter=',')
figure(figsize=(10,5))
x=flight[:,0]
y=flight[:,1]
scatter(x,y)
# Usar objetos de tipo poly1d para representar la regresión
myfit=polyfit(x,y,2)
mypol=poly1d(myfit)
x=linspace(0,1200,40)
plot(x, mypol(x),'r')
xlim(0,1200)
ylim(0,700)
xlabel('x/m',fontsize=15)
ylabel('y/m',fontsize=15)
title(u'Ejemplo de un regresión cuadrática\n--Movimiento Parabólico--',fontsize=15)
print "Velocidad inicial en x = "+ str(1/sqrt(-myfit[0]*2/9.8)) + " m/s"
print "Velocidad inicial en y = "+ str(myfit[1]/sqrt(-myfit[0]*2/9.8)) + " m/s"
show()
```

```python
from scipy.optimize import curve_fit
freq=2.*pi/365
def azimsalida(x,a,b,c):
    return a + b*cos(freq*(x)) + c*sin(freq*x)
nonlfit=curve_fit(azimsalida, sale[:,0], sale[:,2],p0=(0,0,0))
figure(figsize=(10,5))
dias=linspace(0,365,400)
salidafit=azimsalida(dias,nonlfit[0][0],nonlfit[0][1],nonlfit[0][2])
plot(sale[:,0],sale[:,2],'ro',label='datos')
plot(dias,salidafit,label='fit')
xlim(1,365)
xlabel(u'día del año')
ylabel(r'azimut del sol al salir/$^o$')
show()
```

```python
from scipy.optimize import curve_fit
freq=2.*pi/365
def horsalida(x,a,b,c,d,e):
    return a + b*cos(freq*(x)) + c*sin(freq*x) + d*cos(2*freq*x) + e*sin(2*freq*x)
nonlfit=curve_fit(horsalida, sale[:,0], sale[:,1],p0=(0,0,0,0,0))
figure(figsize=(10,5))
dias=linspace(0,365,400)
salidafit=horsalida(dias,nonlfit[0][0],nonlfit[0][1],nonlfit[0][2],nonlfit[0][3],nonlfit[0][4])
plot(sale[:,0],sale[:,1],'ro',label='datos')
plot(dias,salidafit,label='fit')
xlim(1,365)
xlabel(u'día del año')
ylabel(r'hora del amanecer/h')
show()
```

## Óptica Cuántica

Ejemplo de regresión no lineal.
Datos del laboratorio de Óptica Cuántica: operaciones sobre un qubit
$$\sin ^2(\text{p2}+\text{p3}-x) \sin ^2(2 \text{p1}-\text{p2}-\text{p3}+x)+\cos ^2(\text{p2}+\text{p3}-x)
   \cos ^2(2 \text{p1}+\text{p2}-\text{p3}-x)$$
   p1, p2, y p3 son parámetros

```python
# Ejemplo de regresión no polinomial
# Datos del laboratorio de Óptica Cuántica: operaciones sobre un qubit
quarterdata1=genfromtxt(os.path.expanduser("~/Dropbox/Data/blue.csv"),delimiter=',')
quarterdata2=genfromtxt(os.path.expanduser("~/Dropbox/Data/red.csv"),delimiter=',')
def func(x,p1,p2,p3):
    return (cos(p3 + x - 2*p1 - p2) * cos(p3 - x + p2))**2 + (sin(p3 - x + p2) * sin(p3 - x - 2*p1 + p2))**2
# El último argumento entregado a curve_fit corresponde a los parámetros iniciales
nonlfit=curve_fit(func, quarterdata1[:,0], quarterdata1[:,1],p0=(0,0,0))
figure(figsize=(10,5))
ylabel(r'$P(\theta)$',fontsize=15)
xlabel(r'$\theta$',fontsize=15)
th=linspace(0,pi,50)
plot(quarterdata1[:,0],quarterdata1[:,1],"or")
plot(quarterdata2[:,0],quarterdata2[:,1],"oc")
#plot(th,func(th,0.4237,0.058,-0.1529),"r")
plot(th,func(th,nonlfit[0][0],nonlfit[0][1],nonlfit[0][2]),"r")
plot(th,1-func(th,0.4237,0.058,-0.1529),'b')
xlim(0,pi)
ylim(0,1)
title(u'Ejemplo de regresión no polinómica con datos\n del laboratorio de Óptica Cuántica',fontsize=15)
show()
```

## El cometa 26P / Grigg-Skjellerup

```python
# Ejemplo astronómico
grigg=genfromtxt(os.path.expanduser("~/Dropbox/Data/grigg.csv"),delimiter=',')
```

```python
figure(figsize=(10,5))
plot(grigg[:,0],grigg[:,1],"o")
title(u'Orbita del cometa 26P / Grigg-Skjellerup\n marzo 2013 - octubre 2014')
xlabel('x/AU')
ylabel('y/AU')
show()
```

```python
# Calcular la distancia al sol
r=sqrt(grigg[:,0]**2+grigg[:,1]**2)
# y el ángulo que orienta al cometa en su órbita.
angs=arctan2(grigg[:,1],grigg[:,0])
figure()
plot(angs,r,"o")
xlabel(r'$\theta$')
ylabel('r')
show()
```

```python
HTML('<iframe src=http://en.wikipedia.org/wiki/Orbit_equation width=800 height=350></iframe>')
```

$$r(\theta)=\frac{a}{1+\epsilon\cos\theta}$$

```python
def func(x,p1,p2):
    return p1/(1+p2*cos(x))
nonlfit=curve_fit(func,angs,r,p0=(0,0))
a=nonlfit[0][0]
e=nonlfit[0][1]
angs2=linspace(0,2*pi,100)
radios=func(angs2,a,e)
r=sqrt(grigg[:,0]**2+grigg[:,1]**2)
angs=arctan2(grigg[:,1],grigg[:,0])

figure(figsize=(10,5))
plot(angs,r,"o")
plot(angs2,radios)
xlabel(r'$\theta$')
ylabel('r/AU')
show()

figure(figsize=(10,5))
plot(grigg[:,0],grigg[:,1],"o",label='puntos orbitales conocidos')
plot(radios*cos(angs2),radios*sin(angs2),label=u"órbita proyectada")
plot([0.],[0.],"o",label='sol')
xlabel('x/AU')
ylabel('y/AU')
legend()
show()
```

```python
semaxismay=a/(1-e**2)
print "Semi eje mayor = "+str(semaxismay)[0:4:1]+' AU'
# Tercera ley de Kepler
periodo=semaxismay**(1.5)
print "Periodo orbital = "+ str(periodo)[0:4:1]+' yr'
```

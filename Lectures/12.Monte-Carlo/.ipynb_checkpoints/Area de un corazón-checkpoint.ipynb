```python
x = np.random.rand(10000)*4 - 2
y = np.random.rand(10000)*4 - 3
```

```python
scatter(x,y, alpha = 0.1, s=50, c = "k")
xlim(-2, 2)
ylim(-3, 1)
```

```python
def upperheart(x):
    return sqrt(1-(abs(x)-1)**2)
def lowerheart(x):
    return arccos(1-abs(x))-pi  
```

```python
t = linspace(-2,2,100)
plot(t,upperheart(t) , color = "blue")
plot(t,lowerheart(t), color = "red")
```

```python
index = where((y < upperheart(x)) & (y > lowerheart(x)))
```

```python
xc = x[index]
yc = y[index]
```

```python
4*4*len(xc)/len(x)
```

```python
t = linspace(-2,2,100)
plot(t,upperheart(t) , color = "blue")
plot(t,lowerheart(t), color = "red")
scatter(x,y, c='k', alpha=0.3)
scatter(xc,yc, c='g', alpha=0.2, s=40)
xlim(-2, 2)
ylim(-3, 1)
```

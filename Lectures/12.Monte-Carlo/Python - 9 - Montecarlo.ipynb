## Calculando pi

Volumen de la n-esfera


$V_n(R)=\frac{\pi^{n/2}}{\Gamma(1+n/2)} R^n$

```python
# En 2D
darts=100000
cuenta=0
adentro=[]
afuera=[]
for i in xrange(darts):
    x=2*random.random()-1
    y=2*random.random()-1
    if x**2+y**2<=1:
        cuenta+=1
        adentro.append([x,y])
    else:
        afuera.append([x,y])
print "pi = " + str(4.*cuenta/darts)
```

```python
figure(figsize=(10,10))
plot(zip(*adentro)[0],zip(*adentro)[1],"o",ms=3)
plot(zip(*afuera)[0],zip(*afuera)[1],"ro",ms=3)
show()
```

```python
# En 3D
from mpl_toolkits.mplot3d import Axes3D # Para hacer gráficas 3D
darts=100000
cuenta=0
adentro=[]
afuera=[]
for i in xrange(darts):
    x=2*random.random()-1
    y=2*random.random()-1
    z=2*random.random()-1
    if x**2+y**2+z**2<=1:
        cuenta+=1
        adentro.append([x,y,z])
    else:
        afuera.append([x,y,z])
print "pi = " + str(3./4.*8.*cuenta/darts)
```

```python
figure(figsize=(10,10))
gca(projection='3d')
plot(zip(*adentro)[0],zip(*adentro)[1],zip(*adentro)[2],"oc",ms=5)
plot(zip(*afuera)[0],zip(*afuera)[1],zip(*afuera)[2],"ro",ms=1)
show()
```

```python
# En 5D
darts=100000
ene=5
cuenta=0
vol=2**ene
adentro=[]
afuera=[]
for i in xrange(darts):
    r=random.random(ene)*2-1
    if sum(r*r)<=1:
        cuenta+=1
print "pi = " + str(sqrt(15/8.*vol*cuenta/darts))
```

$$\sigma \propto \frac{1}{\sqrt{N}}$$
¡Independiente de la dimensión de la integral!

Estimar la integral con un 10% de precisión.

$$I=\int_0^1\ldots\int_0^1\frac{\textrm{d}\vec{a}\textrm{d}\vec{b}\textrm{d}\vec{c}}{(\vec{a}+\vec{b})\cdot\vec{c}}$$

```python
varios=100
intentos=1000
historia=[]
for k in xrange(varios):
    suma=0
    for i in xrange(intentos):
        veca=random.random(3)
        vecb=random.random(3)
        vecc=random.random(3)
        suma+=1./(sum((veca+vecb)*vecc))
    historia.append(suma/intentos)
```

```python
print mean(historia)
print std(historia)
print std(historia)/mean(historia)*100
```

## Bingo

N personas juegan Bingo, ¿cuál es la probabilidad de que el ganador se obtenga en el turno n?

```python
# Produce una lista de c números enteros aleatorios en el intervalo [a,b-1] sin que se repitan
def intsample(a,b,c):
    done=0
    nums=[]
    if b-a < c:
        return 0
    while len(nums)<c:
        num=random.randint(a,b)
        if not(num in nums):
            nums.append(num)
    return nums
            
def carton():
    bcol=intsample(1,16,5)
    icol=intsample(16,31,5)
    ncol=intsample(31,46,4)
    gcol=intsample(46,61,5)
    ocol=intsample(61,76,5)
    return bcol+icol+ncol+gcol+ocol
```

```python
# Un solo juego
jugadores=10
gano=0
bolsa=range(1,76) # Lista de números por sacar
juego=[0 for i in range(jugadores)] # Lista que guarda el número de aciertos de cada jugador
# Generar una lista con todos los cartones
cartones=[]
numcart=0
while numcart < 10:
    elcarton=carton()
    if elcarton not in cartones:
        cartones.append(carton())
        numcart+=1
turno=0
while gano==0:
    salio=random.choice(bolsa)
    if len(bolsa)!=0:
        bolsa.remove(salio)
    for k in range(jugadores): # Actualizar el número de aciertos de cada jugador
        if salio in cartones[k]:
            juego[k]+=1
    if 24 in juego:
        gano=1
    turno+=1
```
